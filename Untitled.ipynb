```python
x=df.drop("landslide", axis=1)
y=df["landslide"]
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=42)
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print("\n confusion matrix:\n", confusion_matrix(y_test,y_pred))
```

```python
print("\n Classification Report: \n", classification_report(y_test,y_pred))
```

```python
importances=model.feature_importances_
```

```python
feature=x.columns
indices=np.argsort(importances)
```

```python
plt.figure(figsize=(10,6))
plt.title('Feature Importances')
plt.barh(range(len(indices)), [feature[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python

def assign_risk_label(row):
    if row['rainfall_mm'] < 40 and row['soil_moisture_percent'] < 30:
        return 0  # Safe
    elif row['rainfall_mm'] < 80 and row['soil_moisture_percent'] < 45:
        return 1  # Moderate
    else:
        return 2  # Danger

df['risk_level'] = df.apply(assign_risk_label, axis=1)
```

```python
X = df.drop(['landslide', 'risk_level'], axis=1) 
y = df['risk_level'] 
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix

models = {
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(probability=True),
    "XGBoost": XGBClassifier()
}

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f"\n{name} Results:")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```

```python
params = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5]
}

grid_search = GridSearchCV(RandomForestClassifier(), params, cv=3)
grid_search.fit(x_train, y_train)
print("Best Parameters:", grid_search.best_params_)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import label_binarize

# Binarize labels for ROC
y_bin = label_binarize(y_test, classes=[0,1,2])
y_pred_prob = model.predict_proba(x_test)

# One-vs-Rest ROC Curve for each class
for i in range(3):
    fpr, tpr, _ = roc_curve(y_bin[:, i], y_pred_prob[:, i])
    plt.plot(fpr, tpr, label=f"Class {i}")
    
plt.title("Multiclass ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()

# AUC
auc = roc_auc_score(y_bin, y_pred_prob, multi_class='ovr')
print("AUC Score:", auc)
```
